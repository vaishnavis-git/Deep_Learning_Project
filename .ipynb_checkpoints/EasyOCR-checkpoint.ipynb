```python
import matplotlib.pyplot as plt
import easyocr
import matplotlib.patches as patches
```

```python
reader = easyocr.Reader(['en'])

image_path = 'generateimg.jpg'

image = plt.imread(image_path)

fig, ax = plt.subplots(figsize=(8, 6), dpi=100)
ax.imshow(image)
ax.axis('off') 

plt.show()

result = reader.readtext(image_path)

for detection in result:
    print(detection[1])  
```

```python
for detection in result:
    text = detection[1]  
    bbox = detection[0]  
    print("Text:", text)
    print("Bounding Box:", bbox)
```

```python
import matplotlib.pyplot as plt
import easyocr
import numpy as np



# Create a binary mask image
mask = np.zeros_like(image[:, :, 0], dtype=np.uint8)

# Draw filled bounding boxes on the mask image
for detection in result:
    bbox = detection[0]  # The bounding box coordinates
    
    # Extract bounding box coordinates and convert to integers
    x1, y1 = map(int, bbox[0])  # Top-left corner coordinates
    x2, y2 = map(int, bbox[2])  # Bottom-right corner coordinates
    
    # Draw filled rectangle on the mask image
    mask[y1:y2, x1:x2] = 255  # Set region inside bounding box to white (255)

# Display the mask image
plt.imshow(mask, cmap='gray')
plt.axis('off')
plt.show()
```
